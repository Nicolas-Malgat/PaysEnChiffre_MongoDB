```python
import pandas as pd
from pandas.core.series import Series
pays = pd.read_csv("../datas/RAW/25d9c746-3622-4c48-835e-d7ccafa311f5.csv")
len(pays)
```

```python
pays = pays.dropna()
len(pays)
```

```python
def clean_row(col):
    col_values = []

    for value in col:
        if type(value) != str:
            col_values.append(value)
            continue

        value = value.replace('%', '').strip()

        if value.lower().islower(): # verifie que la value contient du texte
            # if value == "N.A.":
            #     col_values.append(0)
            continue
        else:
            try:
                value = int(value)          # verifie si la value contient un int
            except ValueError as e:
                value = float(value)        # la ligne contient un float

        col_values.append(value)
    return Series(col_values)

pays = pays.apply(clean_row)
```
